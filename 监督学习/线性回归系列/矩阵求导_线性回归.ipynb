```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
boston = datasets.load_boston()

X = boston.data
y = boston.target

X = X[y < 50.0]
y = y[y < 50.0]
X.shape
```

```python
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(X, y)
```

```python
import numpy as np
class LinearReg:
    def __init__(self):
        self.coef=None
        self.intercept=None
        self.theta=None
    def fit(self,trainX,trainY):

        X_b = np.hstack([np.ones((len(trainX), 1)), trainX])
        self.theta=np.linalg.inv(np.dot(X_b.transpose(),X_b)).dot(X_b.transpose()).dot(trainY)
        self.intercept=self.theta[0]
        self.coef=self.theta[1:]
    def predict(self,testX):
        X_b = np.hstack([np.ones((len(testX), 1)), testX])
        return np.dot(self.theta,X_b.T)
    
    
```

```python
# 测试线性回归
lr=LinearReg()
lr.fit(trainX,trainY)
lr.predict(testX)
```

```python
# sklean 线性回归
from sklearn.linear_model import LinearRegression
l=LinearRegression();
l.fit(trainX,trainY)
print(l.coef_)
print(l.intercept_)
l.predict(testX)
```
